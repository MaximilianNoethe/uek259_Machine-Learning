```python
from transformers import pipeline, set_seed
```

```python
generator = pipeline('text-generation', model='openai-community/gpt2')

generator("Today is Wednesday",
          truncation = True,
          num_return_sequences=2)
```

```python
generator("Noser Young is a,",
          truncation = True,
          max_length=3,
          num_return_sequences=1)
```

```python
set_seed(2)

generator("Hello, I'm a language model,",
          truncation = True,
          max_length=1,
          num_return_sequences=1)
```

# Sentiment Analysis

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier("I loved Noser Young so much!")
```

```python
pipeline(task = "sentiment-analysis", model = "facebook/bart-large-mnli")("I loved Noser Young so much!")
```

```python
pipeline(task = "sentiment-analysis", model = "facebook/bart-large-mnli")("Noser Young is a, has been working in the NFL for years now and the Seahawks are the only team he\'s ever played for.")
```

```python
pipeline(task = "sentiment-analysis", model = "SamLowe/roberta-base-go_emotions")("I loved Noser Young so much!")
```
